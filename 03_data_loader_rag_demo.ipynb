## With Chat Model

```python
from langchain_openai import ChatOpenAI

chat_model = ChatOpenAI(model="gpt-3.5-turbo-0125")
```

```python
messages = [
    ("system", "You are an historian expert in the history of the Roman Empire."),
    ("human", "Tell me about the fall of the Roman Empire.")
]

response = chat_model.invoke(messages)
print(response.content)
```

```python
messages = [
    ("system", "You are a Geopolitics expert."),
    ("human", "Tell me about the tarriff imposed by US on China in 2025.")
]

response = chat_model.invoke(messages)
print(response.content)
```

```python
# wikipedia data loading
from langchain_community.document_loaders import WikipediaLoader

loader = WikipediaLoader(query="Tesla", load_max_docs=1)
loaded_data = loader.load()[0].page_content
loaded_data
```

```python
# pass the loaded data to the chat model as context to perform question answering
from langchain_community.document_loaders import WikipediaLoader
from langchain_core.prompts import ChatPromptTemplate

loader = WikipediaLoader(query="Tesla", load_max_docs=1)
loaded_data = loader.load()[0].page_content


chat_template = ChatPromptTemplate.from_messages(
    [
        ("human", "Answer this {question}, here is some extra {context}")
    ]
)

messages = chat_template.format_messages(
    name="Tesla",
    question="Where was he born?",
    context=loaded_data
)

response = chat_model.invoke(messages)
response.content
```

#### Drawbacks:
* Input limit issue (If the number of tokens in the document is more than the context window size of the GPT model, then we will not be able to feed it to the model. Check the details of context window & max output tokens of GPT-3.5 Turbo model: https://platform.openai.com/docs/models/gpt-3.5-turbo)
* Higher Cost (As the number of tokens are more, it will be costly for us. Check the pricing: https://platform.openai.com/docs/pricing)

**Is there any way to overcome these drawbacks?**

Yes, with the help of Retreiver Augmented Generation (RAG).
